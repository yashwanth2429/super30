```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("zomato_data.csv")
df.head()
```

```python
print(df)
```

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

df['rate']=df['rate'].apply(handleRate)
print(df.head())
```

```python
df['rate'].mean()
```

```python
df.info()
```

```python
#type of restaurant
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("type of restaurant")
```

```python
grouped_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c='green',marker='o')
plt.xlabel('Type of Restaurant',c='red',size=20)
plt.ylabel('Votes',c='red',size=20)
```

```python
df.head()
```

```python
plt.hist(df['rate'],bins=15)
plt.title('Ratings Distribution')
plt.show()
```

```python
#average spendings by couples
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=df)
```

```python
pivot_table=df.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title("Heatmap")
plt.xlabel('Online Order')
plt.ylabel('listed_in(type)')
plt.show()
```

**Conclusion:dining restaurants primarily accept offline orders,whereas cafes primarily receive online orders.This suggests that clients prefers orders in person at restaurants, but prfer online orders at cafes**
